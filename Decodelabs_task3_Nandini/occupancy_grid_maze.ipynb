```python
from queue import PriorityQueue

def astar(start, goal):
    pq = PriorityQueue()
    pq.put((0,start))
    visited = set()

    while not pq.empty():
        cost,node = pq.get()

        if node == goal:
            print("Goal Reached")
            return

        visited.add(node)

        for neighbor in [(node[0]+1,node[1]),
                         (node[0]-1,node[1]),
                         (node[0],node[1]+1),
                         (node[0],node[1]-1)]:

            if neighbor not in visited:
                pq.put((cost+1,neighbor))

astar((0,0),(4,4))
```
