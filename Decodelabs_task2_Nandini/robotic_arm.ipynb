```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Arm lengths
L1 = 2
L2 = 2

# Target
x = 2
y = 2

# Inverse Kinematics
d = (x**2 + y**2 - L1**2 - L2**2)/(2*L1*L2)
theta2 = np.arccos(d)

theta1 = np.arctan2(y,x) - np.arctan2(
    L2*np.sin(theta2),
    L1 + L2*np.cos(theta2)
)

# Joint coordinates
x1 = L1*np.cos(theta1)
y1 = L1*np.sin(theta1)

x2 = x1 + L2*np.cos(theta1+theta2)
y2 = y1 + L2*np.sin(theta1+theta2)

# Plot
plt.plot([0,x1,x2],[0,y1,y2],
         marker='o',
         linewidth=4)
plt.scatter(x2,y2,s=150)
plt.text(x2,y2," End Effector")
plt.scatter(0,0,s=100)
plt.text(0,0," Base")

plt.scatter(x1,y1,s=100)
plt.text(x1,y1," Joint 1")
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid()
plt.title("Robotic Arm Simulator")
plt.show()
```
